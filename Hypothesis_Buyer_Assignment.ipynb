```python
chiStats =stats.chi2_contingency(Buyers)
chiStats
```

```python
chiStats[1]
print('P_value is:',chiStats[1])
if chiStats[1] < 0.05:
  print('we reject null hypothesis')
else:
  print('we accept null hypothesis')
```
